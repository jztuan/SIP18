```python
gds = Table.read(data_dir+'gds.hdf5')
```

```python
gds_selection = np.where ( \
                          (gds['PhotFlag']==0) &\
                          (gds['CLASS_STAR']>0.9) & \
                          (gds['ACS_F606W_FLUX'] > 0) &\
                         (gds['ACS_F814W_FLUX'] > 0) &\
                          (gds['WFC3_F125W_FLUX'] > 0) &\
                         (gds['WFC3_F160W_FLUX'] > 0))

# last 3 conditions ensure log10() exists
```

```python
gds_temp = gds[gds_selection]
gds_selection2 = np.where ( \
                          -2.5*np.log10(gds_temp['WFC3_F125W_FLUX']) <= 1.359465)
gds_final = gds_temp[gds_selection2]
```

```python
## define variables for GOODS-S

f606_gds_flux = gds_final['ACS_F606W_FLUX']
f814_gds_flux = gds_final['ACS_F814W_FLUX']  ## rest-frame .814 μm magnitude
f125_gds_flux = gds_final['WFC3_F125W_FLUX']   ## rest-frame 1.25 μm magnitude
f160_gds_flux = gds_final['WFC3_F160W_FLUX']   ## rest-frame 1.60 μm magnitude

f606_gds_mag = -2.5*np.log10(f606_gds_flux)
f814_gds_mag = -2.5*np.log10(f814_gds_flux)
f125_gds_mag = -2.5*np.log10(f125_gds_flux)
f160_gds_mag = -2.5*np.log10(f160_gds_flux)

f606_f814_gds = f606_gds_mag - f814_gds_mag
f814_f125_gds = f814_gds_mag - f125_gds_mag
f125_f160_gds = f125_gds_mag - f160_gds_mag

# mstar_gds = gds_final['M_med'] # already logged?
# z_gds = gds_final['z_best']  ## redshift
```

```python
## define variables for GOODS-S

f606_cos_flux = cos_final['ACS_F606W_FLUX']
f814_cos_flux = cos_final['ACS_F814W_FLUX']  ## rest-frame .814 μm magnitude
f125_cos_flux = cos_final['WFC3_F125W_FLUX']   ## rest-frame 1.25 μm magnitude
f160_cos_flux = cos_final['WFC3_F160W_FLUX']   ## rest-frame 1.60 μm magnitude

f606_cos_mag = -2.5*np.log10(f606_cos_flux)
f814_cos_mag = -2.5*np.log10(f814_cos_flux)
f125_cos_mag = -2.5*np.log10(f125_cos_flux)
f160_cos_mag = -2.5*np.log10(f160_cos_flux)

f606_f814_cos = f606_cos_mag - f814_cos_mag
f814_f125_cos = f814_cos_mag - f125_cos_mag
f125_f160_cos = f125_cos_mag - f160_cos_mag

# mstar_cos = cos_final['M_med'] # already logged?
# z_cos = cos_final['z_best']  ## redshift
```

```python
f606_f814 = np.concatenate([f606_f814_gds,f606_f814_egs,f606_f814_gdn,f606_f814_cos])
f814_f125 = np.concatenate([f814_f125_gds,f814_f125_egs,f814_f125_gdn,f814_f125_cos])
f125_f160 = np.concatenate([f125_f160_gds,f125_f160_egs,f125_f160_gdn,f125_f160_cos])
total_length = len(f606_f814)
print total_length
```

```python
f814_f125_sorted1 = [x for _,x in sorted(zip(f606_f814,f814_f125))]
f606_f814_sorted1 = np.sort(f606_f814)
movingmedian_f606_f8141 = np.array([])
movingmedian_f814_f1251 = np.array([])
for i in range (0,total_length-15):
    data2 = np.array([])
    for j in range(i,i+15):
        data2 = np.append(data2,f606_f814_sorted1[j])
    movingmedian_f606_f8141 = np.append(movingmedian_f606_f8141,np.median(data2))
for k in range (0,total_length-15):
    data3 = np.array([])
    for l in range(k,k+15):
        data3 = np.append(data3,f814_f125_sorted1[l])
    movingmedian_f814_f1251 = np.append(movingmedian_f814_f1251,np.median(data3))
# print movingmedian_f606_f814
```

```python
fig=plt.figure(figsize=(8,8))
plt.scatter(f814_f125_sorted1,f606_f814_sorted1,c='grey',edgecolors='none',s=8,alpha=0.7)
plt.scatter(movingmedian_f814_f1251,movingmedian_f606_f8141,c='blue',edgecolors='none',s=8,alpha=0.7)
plt.xlim(-0.5,4)
plt.ylim(-0.5,4)
plt.savefig('/Users/Andy/Desktop/GIJ_pt4_starlike_movingmedian.pdf')
```

```python
distance_606_814_125 = np.array([])
for m in range(0,total_length-15):
    min_distance = 99999
    for n in range(0,total_length-15):
        dist = math.sqrt((movingmedian_f606_f8141[n]-f606_f814_sorted1[m])**2+(movingmedian_f814_f1251[n]-f814_f125_sorted1[m])**2)
        if (dist < min_distance):
            min_distance = dist
    distance_606_814_125 = np.append(distance_606_814_125,min_distance)
print distance_606_814_125
keep = np.where(distance_606_814_125 < 1)
plt.hist(distance_606_814_125[keep],100,color='orange')
```
